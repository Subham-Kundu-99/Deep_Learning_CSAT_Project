## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# RELOAD THE DATAFRAME TO ENSURE A CLEAN START
df = pd.read_csv("/content/eCommerce_Customer_support_data.csv")

# Standardizing column names
df.columns = df.columns.str.strip().str.replace(" ", "_")
# Removing duplicates
df = df.drop_duplicates()
#Converting Date Columns
date_cols = [
    "order_date_time",
    "Issue_reported_at",
    "issue_responded",
    "Survey_response_Date"

]
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
#Time Features
#Resolution Time (in minutes)
df["resolution_time_minutes"] = (
    (df["issue_responded"] - df["Issue_reported_at"])
    .dt.total_seconds() / 60
)
#Order to Issue Delay
df["order_to_issue_hours"] = (
    (df["Issue_reported_at"] - df["order_date_time"])
    .dt.total_seconds() / 3600
)
#Extracting Day & Month
df["order_month"] = df["order_date_time"].dt.month
df["order_day"] = df["order_date_time"].dt.day


#Handling Missing Values
#Filling numerical columns with median
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Filling categorical columns with mode, excluding 'Customer_Remarks' for later text processing
text_column_for_nlp = 'Customer_Remarks'

categorical_cols_for_label_encoding = [col for col in df.select_dtypes(include='object').columns if col != text_column_for_nlp]

for col in categorical_cols_for_label_encoding:
    df[col] = df[col].fillna(df[col].mode()[0])

# Handle missing for Customer_Remarks separately, e.g., fill with empty string
if text_column_for_nlp in df.columns:
    df[text_column_for_nlp] = df[text_column_for_nlp].fillna('')

#Encode Categorical Variables
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in categorical_cols_for_label_encoding: # Only encode the non-text columns
    df[col] = le.fit_transform(df[col])

#Drop Unnecessary Columns
drop_cols = [
    "Unique_id",
    "Order_id",
    # "Customer_Remarks", # Do not drop this if we want to vectorize it
    "Agent_name"
]

df = df.drop(columns=drop_cols, errors='ignore')
#Final Dataset Info
print("Dataset Shape:", df.shape)
print("\nMissing Values Remaining:\n", df.isnull().sum().sum())
print("\nDataset Ready for ML")
````

The dataset was prepared for analysis through a structured sequence of cleaning, transformation, and feature engineering steps to make it suitable for exploratory data analysis and machine learning modeling. First, column names were standardized by removing spaces and formatting them consistently to avoid errors during processing. Duplicate records were removed to ensure data integrity. Date-related columns such as order date, issue reporting time, response time, and survey response date were converted into proper datetime format to enable time-based analysis.

Feature engineering was performed to derive meaningful operational metrics. Mean imputation was applied to numerical columns because replacing missing values with the average helps maintain the overall distribution of the data without introducing extreme values. Mode imputation was used for categorical columns because the mode represents the most frequently occurring category, making it a suitable replacement for missing categorical data. These techniques are simple, effective, and commonly used during data pre-processing to ensure the dataset remains complete for further analysis.

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
import numpy as np
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

print("Outliers handled successfully")
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

The Interquartile Range (IQR) method was used for outlier detection and treatment. This technique identifies outliers by calculating the first quartile (Q1), third quartile (Q3), and the interquartile range (IQR). Any values that fall below Q1 − 1.5 × IQR or above Q3 + 1.5 × IQR are considered outliers.

This method was chosen because it is simple, effective, and does not get affected by extreme values. It works well for numerical data and helps remove unusual observations that could negatively impact the accuracy of data analysis and statistical results.

#### 3. Removing Punctuations

````python
# Remove Punctuations
import string
def remove_punctuation(text):
    if isinstance(text, str):
        return text.translate(str.maketrans('', '', string.punctuation))
    return text
text_cols = df.select_dtypes(include=['object']).columns

for col in text_cols:
    df[col] = df[col].apply(remove_punctuation)

print("Punctuations removed successfully")
````

#### 4. Removing URLs & Removing words and digits contain digits.

````python
# Remove URLs & Remove words and digits contain digits
import re
def remove_urls(text):
    if isinstance(text, str):
        return re.sub(r'http\S+|www\S+', '', text)
    return text
def remove_words_with_digits(text):
    if isinstance(text, str):
        return re.sub(r'\w*\d\w*', '', text)
    return text
text_cols = df.select_dtypes(include=['object']).columns
for col in text_cols:
    df[col] = df[col].apply(remove_urls)
    df[col] = df[col].apply(remove_words_with_digits)

print("URLs and words containing digits removed successfully")
````

#### 5. Removing Stopwords & Removing White spaces

````python
# Remove Stopwords
import nltk
from nltk.corpus import stopwords

# Download stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

# Function to remove stopwords
def remove_stopwords(text):
    if isinstance(text, str):
        words = text.split()
        filtered_words = [word for word in words if word not in stop_words]
        return " ".join(filtered_words)
    return text

# Apply only on text columns
text_cols = df.select_dtypes(include=['object']).columns

for col in text_cols:
    df[col] = df[col].apply(remove_stopwords)

print("Stopwords removed successfully")
````

````python
# Remove White spaces
text_cols = df.select_dtypes(include=['object']).columns
for col in text_cols:
    if not df[col].empty and isinstance(df[col].iloc[0], str):
        df[col] = df[col].str.strip()
        df[col] = df[col].str.replace(r'\s+', ' ', regex=True)
    else:
        print(f"Skipping white space removal for column '{col}' as it does not contain string values.")
print("White spaces removed successfully")
````

#### 6. Rephrase Text

````python
# Rephrase Text
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
lemmatizer = WordNetLemmatizer()
def lemmatize_text(text):
    if isinstance(text, str):
        words = text.split()
        words = [lemmatizer.lemmatize(word) for word in words]
        return " ".join(words)
    return text
text_cols = df.select_dtypes(include=['object']).columns
for col in text_cols:
    df[col] = df[col].apply(lemmatize_text)

print("Text rephrased successfully")
````

#### 7. Tokenization

````python
# Tokenization
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
def tokenize_text(text):
    if isinstance(text, str):
        return word_tokenize(text)
    return text
text_cols = df.select_dtypes(include=['object']).columns

for col in text_cols:
    df[col] = df[col].apply(tokenize_text)

print("Tokenization completed successfully")
````

#### 8. Text Normalization

````python
# Normalizing Text (i.e., Stemming, Lemmatization etc.)
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Download required resources
nltk.download('wordnet')
nltk.download('omw-1.4')

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

def normalize_text(text):
    if isinstance(text, str):
        words = text.split()
        stemmed_words = [stemmer.stem(word) for word in words]
        lemmatized_words = [lemmatizer.lemmatize(word) for word in stemmed_words]
        return " ".join(lemmatized_words)
    return text

# Apply to text columns
text_cols = df.select_dtypes(include=['object']).columns

for col in text_cols:
    df[col] = df[col].apply(normalize_text)

print("Text normalization completed successfully")
````

##### Which text normalization technique have you used and why?

The text normalization techniques used are Stemming and Lemmatization. Stemming reduces words to their root form by removing suffixes (for example, running becomes run). Lemmatization converts words into their meaningful base form based on vocabulary and grammar (for example, better becomes good). These techniques were used to standardize different forms of words into a common base form, which helps reduce data complexity and improves the effectiveness of text analysis.

#### 9. Part of speech tagging

````python
# POS Taging
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
# Function for POS tagging
def pos_tagging(text):
    if isinstance(text, str):
        tokens = word_tokenize(text)
        return pos_tag(tokens)
    return text
text_cols = df.select_dtypes(include=['object']).columns

for col in text_cols:
    df[col] = df[col].apply(pos_tagging)

print("POS tagging completed successfully")
````

#### 3. Explain each evaluation metric's indication towards business and the business impact pf the ML model used.

The R² score shows how well the model explains the variation in the target variable. A higher R² value means the model can better understand the relationship between input features and the target. From a business perspective, a high R² indicates that the predictions are reliable, which helps companies make better data-driven decisions such as improving customer service operations, predicting support performance, or optimizing resources.

* **Mean Squared Error (MSE)**:
MSE measures the average squared difference between the actual values and the predicted values. A lower MSE means the prediction errors are small. In a business context, lower error means the model’s predictions are closer to reality, which reduces the risk of wrong decisions. For example, more accurate predictions can help businesses allocate staff, improve response time, and maintain better service quality.

* **Train vs Test Performance**:
Comparing training and testing scores helps determine whether the model is overfitting or generalizing well. If both scores are similar, it means the model performs consistently on new data. For business use, this is important because it shows that the model can be trusted to work on future real-world data, not just the historical dataset used for training.

* **Overall Business Impact**:
A well-performing ML model with high R² and low MSE helps businesses gain meaningful insights from customer support data. It can support better operational planning, improve customer satisfaction, and help management make more accurate strategic decisions based on data-driven predictions.
